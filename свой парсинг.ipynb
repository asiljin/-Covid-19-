```python
%%time
from datetime import datetime, timedelta
target2=[]
x='https://mix.tn.kz/mixnews'
for i in tqdm(range(len(links2))):
    article_page = requests.get(links2[i])
    article_soup = bs(article_page.text, 'html.parser')

    ##Получение Название
    for name_link in article_soup.find_all('h1',class_='h1'):
        name_link = name_link.text
        name_link = name_link.replace('\n', ' ')
    #print(name_link)
    #print(name)

    ##Получение даты
    for data_link in article_soup.find_all('div',class_='date'):
        right_data = data_link.text
        right_data = right_data.replace('\n', ' ')
        #print(right_data)
        #print(name)
            
    ###получение текста
    text = []
    for text_link in article_soup.find_all('div', class_='desc js-mediator-article'):
            text.append(text_link.text)
            #print(view_link.text)

    ##text.remove(comm[0])
    text=text[0]
    text = text.replace('\n', ' ')
    text = text.replace('\xa0', ' ')
    text
        
    target2.append({ 'name':name_link,
                        'date':right_data,
                        'text':text
                            })
```

```python
%%time
from datetime import datetime, timedelta
target3=[]
x='https://mix.tn.kz/mixnews'
for i in tqdm(range(len(links3))):
    article_page = requests.get(links3[i])
    article_soup = bs(article_page.text, 'html.parser')

    ##Получение Название
    for name_link in art_soup.find_all('h1',class_='tr-single-title'):
        name_link = name_link.text
    #print(name)
        
        
    ##Получение даты
    for data_link in art_soup.find_all('div',class_='tr-single-metadata'):
            right_data = data_link.text
            right_data = right_data[:-2]
            right_data = right_data.replace('\n', ' ')
            #print(right_data)
    #print(name)
        
    text = []
    for text_link in art_soup.find_all('div', class_='tr-single-text'):
        text.append(text_link.text)
        #print(view_link.text)


    text = text[0]
    text = text.replace('\n', ' ')
    text = text.replace('\xa0', ' ')
        
    target3.append({ 'name':name_link,
                        'date':right_data,
                        'text':text
                            })
```
